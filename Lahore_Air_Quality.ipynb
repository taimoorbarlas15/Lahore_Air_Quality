```python
# Outlier Detection using the IQR method for AQI and temperature columns
outlier_columns = ['aqi_pm2.5', 'max_temp_f', 'avg_temp_f', 'min_temp_f']

def detect_outliers(data, columns):
    outlier_counts = {}
    outlier_indices = {}
    for column in columns:
        Q1 = data[column].quantile(0.25)
        Q3 = data[column].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        outliers = data[(data[column] < lower_bound) | (data[column] > upper_bound)]
        outlier_counts[column] = len(outliers)
        outlier_indices[column] = outliers.index.tolist()
    return outlier_counts, outlier_indices

# Detect outliers
outlier_counts, outlier_indices = detect_outliers(df_aqi_updated, outlier_columns)

outlier_counts
```

```python
# Function to handle outliers by capping them at the IQR bounds
def cap_outliers(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    data[column] = np.where(data[column] < lower_bound, lower_bound, data[column])
    data[column] = np.where(data[column] > upper_bound, upper_bound, data[column])
    return data
```

```python
# Apply capping to temperature columns with detected outliers
temperature_columns = ['max_temp_f', 'min_temp_f']
for column in temperature_columns:
    df_aqi_updated = cap_outliers(df_aqi_updated, column)
```

```python
#Insights from Winter vs. Other Seasons Weather Analysis

# Reload the dataset and prepare for analysis if not already available
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import ttest_ind

# Ensure the AQI dataset is available
df_aqi_updated = pd.read_csv("/content/lahore_aqi_2019_to_2023.csv")

# Ensure 'date' is in datetime format
df_aqi_updated['date'] = pd.to_datetime(df_aqi_updated['date'], format='%d-%m-%y', errors='coerce')

# Add month and season columns
df_aqi_updated['month'] = df_aqi_updated['date'].dt.month
df_aqi_updated['season'] = df_aqi_updated['month'].apply(
    lambda x: 'Winter' if x in [12, 1, 2] else 'Other'
)

# Separate weather factors for winter vs. other seasons
weather_factors = ['max_temp_f', 'avg_temp_f', 'min_temp_f', 'max_humidity_percent', 'avg_humidity_percent', 'max_wind_speed_mph']
winter_weather = df_aqi_updated[df_aqi_updated['season'] == 'Winter'][weather_factors]
other_weather = df_aqi_updated[df_aqi_updated['season'] == 'Other'][weather_factors]

# Perform t-tests for each weather factor
weather_stats = {}
for factor in weather_factors:
    t_stat, p_value = ttest_ind(winter_weather[factor], other_weather[factor], equal_var=False)
    weather_stats[factor] = {
        'Winter Mean': winter_weather[factor].mean(),
        'Other Seasons Mean': other_weather[factor].mean(),
        'T-statistic': t_stat,
        'P-value': p_value
    }

# Visualize mean differences for each weather factor
means = pd.DataFrame({
    'Winter': winter_weather.mean(),
    'Other Seasons': other_weather.mean()
}).reset_index()

plt.figure(figsize=(12, 6))
plt.bar(means['index'], means['Winter'], label='Winter', alpha=0.7)
plt.bar(means['index'], means['Other Seasons'], label='Other Seasons', alpha=0.7)
plt.title('Mean Comparison of Weather Factors: Winter vs. Other Seasons')
plt.ylabel('Mean Values')
plt.xticks(rotation=45)
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# Display the t-test results
weather_stats
```

Insights from Winter vs. Other Seasons Weather Analysis

Temperature (Max, Avg, Min):

Winter Mean: Significantly lower than other seasons.

Max Temp: ~66.8°F (Winter) vs. ~91.9°F (Other Seasons)

Avg Temp: ~56.1°F (Winter) vs. ~82.0°F (Other Seasons)

Min Temp: ~45.3°F (Winter) vs. ~70.2°F (Other Seasons)

T-test Results: Extremely low p-values (<0.001) indicate statistically significant differences in temperatures between winter and other seasons.

Implication: Colder winter temperatures contribute to reduced atmospheric mixing, which can trap pollutants near the ground.

Wind Speed:

Winter Mean: Lower than in other seasons.
Max Wind Speed: ~10.4 mph (Winter) vs. ~17.5 mph (Other Seasons)

T-test Results: Statistically significant (p < 0.001).

Lower wind speeds in winter reduce pollutant dispersion, allowing pollutants to accumulate in stagnant conditions.

Explanation of the Regression Statistics

he regression model outputs coefficients for each factor, which quantify their respective impact on PM2.5 (smog) levels

Temperature (°F): Coefficient = -2.717338

A negative coefficient suggests that as the temperature increases, PM2.5 levels decrease.
This is likely because higher temperatures can promote vertical dispersion of air pollutants, reducing their concentration near the ground.
The magnitude (-2.717) indicates that for each 1°F increase in average temperature, PM2.5 levels decrease by approximately 2.72 units, assuming other factors remain constant.

Humidity (%): Coefficient = -0.821906

A negative coefficient indicates that higher humidity is associated with slightly lower PM2.5 levels.
Increased humidity can lead to particle aggregation and deposition, which removes particulate matter from the air.
However, the effect is relatively weak

Wind Speed (mph): Coefficient = -9.523677

A strong negative coefficient implies that higher wind speeds significantly reduce PM2.5 levels.
Wind helps disperse air pollutants over a larger area, reducing their concentration in any specific location.
For each 1 mph increase in wind speed, PM2.5 levels decrease by approximately 9.52 units, showing wind is a critical factor in air quality improvement.


Population: Coefficient = 0.000045

A positive coefficient suggests that as population increases, PM2.5 levels also increase slightly.The effect is small because the coefficient (0.000045) is low. This is likely due to population being aggregated yearly

Vehicles: Coefficient = -0.000038

A negative coefficient here might seem counterintuitive since vehicles are a known source of PM2.5

Insights

Meteorological Factors (temperature, humidity, wind speed) have significant effects on PM2.5, with wind speed being the most impactful.

Non-Meteorological Factors like population and vehicles have weaker statistical contributions in this dataset, potentially due to data aggregation or limitations.
